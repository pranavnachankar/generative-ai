### RAG (Retrieval-Augmented Generation) :



- Load :
    - It is a data source which can consist multiple files like PDF, Img, JSON, etc,..
    - this data we are ingeting for further process
    
- Split :
    - after having the huge data we ar diving this corpus into chunks / docs / sentences

- Embed :
    - Data --->>> Vectors
    
- Store :
    - after converting into "vectors", we need to store this data that we called as "Vector Store database"
    - it usally use Similarity search algo

- Question : a sample qtns asked by the user which will get combinely applied with `prompt`
- Prompt : it is loaded with few system instructions that tells bot how to behave
- `Question + Prompt` will combinly apply to the database to retrive the data
- now `Question + Prompt` internally use `Retrival chain` and trigger the query to DB
- after that we will get `Context information` from DB
- then `Question + Prompt` + `Context information` will apply to `LLM`
- `LLM` will process the data and give `meaningful information`.
